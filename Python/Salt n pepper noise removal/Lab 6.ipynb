```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('2.png', 0)

kernel = np.ones((3,3), np.float32)/9
dst = cv2.filter2D(img, -1, kernel)

result = np.hstack((img, dst))
cv2.imshow('result', result)

cv2.waitKey(0)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('4.png', 1)

blur = cv2.blur(img, (5,5)) #src, ksize

result = np.hstack((img, blur))
cv2.imshow('result', result)

cv2.waitKey(0)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('4.png', 1)

box = cv2.boxFilter(img, -1, (5,5)) #src, depth, ksize

result = np.hstack((img, box))
cv2.imshow('result', result)

cv2.waitKey(0)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('1.png', 1)

median = cv2.medianBlur(img, 3) #src, ksize

result = np.hstack((img, median))
cv2.imshow('result', result)

cv2.waitKey(0)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('1.png', 0)

gaussian = cv2.GaussianBlur(img, (11, 11), 0) #src, ksize, standard deviation sigmaX

result = np.hstack((img, gaussian))
cv2.imshow('result', result)

cv2.waitKey(0)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

hubble = cv2.imread('3.png', 0)

kernel = np.ones((15,15), np.float32)/225
dst = cv2.filter2D(hubble, -1, kernel)
 
res,thresh = cv2.threshold(dst, 120, 255, cv2.THRESH_BINARY)

result = np.hstack((hubble, dst, thresh))
cv2.imshow('result', result)

cv2.waitKey(0)

plt.imshow(cv2.cvtColor(hubble, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
plt.show()
```
